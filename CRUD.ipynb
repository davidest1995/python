```python
import csv
import os
from typing_extensions import TypeGuard


USER_TABLE='/.users.csv'
USER_SCHEMA=['Name','Email','Password','Document','Investment']
users = []

def _initialaze_users_from_storage():
  with open(USER_TABLE, mode='r') as f:
      reader = csv.DictReader(f,fieldnames=USER_SCHEMA)
      for row in reader:
        users.append(row)

def _save_users_to_storage():
  tmp_table_name='{}.tmp'.format(USER_TABLE)
  with  open(tmp_table_name, mode='w') as f:
    writer=csv.DictWriter(f,fieldnames=USER_SCHEMA)
    writer.writerows(users)

    os.remove(USER_TABLE)
    os.rename(tmp_table_name, USER_TABLE)  


def user_not_found():
  print('El usuario no se ha encontrado')

def list_users():
  print('uid  |  Name  |  Email  |  Password  |  Document  |  Investment ')
  print('*'* 50)
  for idx, user in enumerate(users):
    print('{uid}  |  {Name}  |  {Email}  |  {Password}  |  {Document}  |  {Investment} '.format(
        uid=idx, 
        Name = user['Name'], 
        Email = user['Email'], 
        Password = user['Password'], 
        Document = user['Document'], 
        Investment = user['Investment']))

def create_user(user):
  global users
  if user not in users:
    users.append(user)
  else:
    print('User is already in the user\'s list')

def _get_user_field(field_name,message='What is the user\'s {} '):
  field = None
  while not field:
    field = input(message.format(field_name)) 

  return field  
def _get_user_name():
  user_name = None
  while not user_name:
      user_name = input('Ingrese nombre completo: ')
      if user_name == 'exit':
        user_name = None
        break
  if not user_name:
    raise SystemExit('El programa se detuvo')
  return user_name

def delete_user(user_id):
  global users
  for idx,user in enumerate(users):
    if idx == user_id:
      del users[idx]
      break

def _get_user_from_users():
    user = {
        'Name': _get_user_field('Name'),
        'Email': _get_user_field('Email'),
        'Password': _get_user_field('Password'),
        'Document': _get_user_field('Document'),
        'Investment': _get_user_field('Investment')
    }
    return user

def update_user(user_id,updated_user):
  global users
  if len(users) >= user_id:
      users[user_id]= updated_user
  else:
    user_not_found()

def search_user(user_name):
  global users
  for user in users:
    if user['Name'] != user_name:
      continue   
    else:
      return True

def _print_welcome():
  print('WELCOME TO ILUMINATY')
  print('*'*50)
  print('What would you like to do today? ')
  print('[L]ist of users')
  print('[C]reate new user')
  print('[U]pdate user')
  print('[D]elete user')
  print('[S]earch user')

if __name__ == '__main__':
  _initialaze_users_from_storage()
  _print_welcome()

  command = input()
  command = command.upper()
  if command == 'C':
    user = _get_user_from_users()
    create_user(user)
    list_users()
  elif command == 'D':
    list_users()
    user_id= int(_get_user_field('id'))
    delete_user(user_id)
    list_users()

  elif command == 'U':
    list_users()
    user_id = int(_get_user_field('id'))
    updated_user = _get_user_from_users()
    update_user(user_id,updated_user)
    list_users()
  elif command == 'L':
    list_users()
  elif command == 'S':
    user_name = _get_user_name()
    found = search_user(user_name)
    if found:
      print('El usuario sí está en la lista id su id es: {}'.format(user_id))
    else:
      print('El usuario {} no está en la lista'.format(user_name))
  else:
    print('Comando Invalido')
  
  _save_users_to_storage()
```
